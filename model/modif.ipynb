```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_science_salaries_2024.csv')
df.head()
```

```python
df.nunique()
```

```python
# look at object values
obj_cols = df.select_dtypes(include='O').columns

for col in obj_cols:
    print(df[col].value_counts(), '\n')
```

```python
df['experience_level'] = df['experience_level'].replace({'Entry-level': 'EN', 'Mid-level': 'MI', 'Senior-level': 'SE', 'Executive-level': 'EX'})
df['employment_type'] = df['employment_type'].replace({'Full-time': 'FT', 'Contract': 'CT', 'Part-time': 'PT', 'Freelance': 'FL'})
df['company_size'] = df['company_size'].replace({'Small': 'S', 'Medium': 'M', 'Large': 'L'})
df['company_location'] = df['company_location'].replace({'United States': 'US'})
df['employee_residence'] = df['employee_residence'].replace({'United States': 'US'})
df['remote_ratio'] = df['work_models'].replace({'On-site': 0, 'Remote': 100, 'Hybrid': 50})
```

```python
df.head()
```

```python
df.to_csv('dsjobs_2024_modified.csv', index=False)
```
